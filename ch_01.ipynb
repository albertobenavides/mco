

## 1.1
Consider the problem $$\text{``min''} (f_1(x), f_2(x)) \text{ subjec to } x \in [-1, 1],$$ where $$f_x(x) = \sqrt{5 - x^2}, f_2(x) = \frac{x}{2}.$$

Illustrate tthe problem in decision and objective space and determine the nondominanted set $\mathcal{Y}_N := {y \in \mathcal{Y}: \text{ there is no } y' \in \mathcal{Y} \text{  with } y' \leq y}$ and the efficient set $\mathcal{X}_E := {x \in \mathcal{X} : f(x) \in \mathcal{Y}_N}$.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,1,100)
f_1 = np.sqrt(5 - np.power(x, 2))
f_2 = x / 2
```

```python
plt.figure()
plt.plot(x, f_1, label = '$f_1$')
plt.plot(x, f_2, label = '$f_2$')
plt.legend()
plt.show()
```

```python
Y_N = (f_1, f_2)
```

```python
plt.figure()
plt.plot(Y_N[0], Y_N[1], label = '$\mathcal{Y}_N$')
plt.legend()
plt.show()
```

- Decision space: $\mathbb{R}$
- Objective space: $f(x) := {y \in \mathbb{R}^2 : y = (f_1(x), f_2(x)) \text{ for } x \in \mathcal(X)}$
- Nondominated set: $\mathcal{Y}_N = (2, -0.5)$
- Efficient set $\mathcal{X}_E = \{-1\}$
